# Ratios Analysis of `Shaheen Air International SAI`

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sai_ratios = pd.read_csv("ratios_sai.csv")
sai_ratios
```

```python
sai_ratios.info()
```

```python
import matplotlib.dates as mdates

# Extract dates and ratios
dates = sai_ratios["Date"]
fixed_asset_turnover = sai_ratios["Fixed Asset Turnover ratio"]
total_asset_turnover = sai_ratios["Total Asset Turnover ratio"]

# Define line coordinates
y_start = 24
y_end = 12

# Extract minimum and maximum dates
min_date = min(dates)
max_date = max(dates)

# Define bar width and spacing
width = 0.25
bar_spacing = 0.3

# Set figure size
fig, ax = plt.subplots(figsize=(12, 6))

# Configure x-axis as dates
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Set x-axis labels
ax.set_xticks(dates)
ax.set_xticklabels(dates, rotation=45, ha="right", fontsize=12)

# Remove unnecessary gridlines and axis outlines
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


# Plot fixed asset turnover bars
ax.bar(
    dates - bar_spacing / 2,
    fixed_asset_turnover,
    width,
    label='Fixed Asset Turnover Ratio',
    color='b',
    alpha=0.8,
    linewidth=1,
    capstyle='round',
)

# Plot total asset turnover bars with offset for spacing
ax.bar(
    dates + bar_spacing / 2,
    total_asset_turnover,
    width,
    label='Total Asset Turnover Ratio',
    color='g',
    alpha=0.8,
    linewidth=1,
    capstyle='round',
)


# Plot line with desired color and style
ax.plot([min_date, max_date], [y_start, y_end], color='teal', linewidth=2, linestyle='--')


# Set y-axis labels
ax.set_ylabel('Ratio', fontweight='bold', fontsize=14,color = "dimgray")

# Add value annotations to each bar
for i, (date, value) in enumerate(zip(dates, fixed_asset_turnover)):
    ax.text(date - bar_spacing / 2, value + 0.05, f"{value:.2f}", ha="center", va="bottom", fontsize=12)

for i, (date, value) in enumerate(zip(dates, total_asset_turnover)):
    ax.text(date + bar_spacing / 2, value + 0.05, f"{value:.2f}", ha="center", va="bottom", fontsize=12)

# Set legend with a frame and rounded corners
legend = ax.legend(loc='upper right', frameon=True, title='Ratios', title_fontsize=12, fancybox=True)

# set title
plt.title("**ASSET MANAGEMENT RATIOS**", fontsize=18, fontweight="bold",color='teal')
# Tight layout and show plot
plt.tight_layout()
plt.show() 

```

```python
import matplotlib.dates as mdates

# Extract dates and ratios
dates = sai_ratios["Date"]
BEP_ratio = sai_ratios["BEP ratio"]
ROA_ratio = sai_ratios["ROA ratio"]
ROE_ratio = sai_ratios["ROE ratio"]

# Define line coordinates
y_start = 77
y_end = -57

# Extract minimum and maximum dates
min_date = min(dates)
max_date = max(dates)

# Define bar width and spacing
width = 0.25
bar_spacing = 0.3

# Set figure size
fig, ax = plt.subplots(figsize=(12, 6))

# Configure x-axis as dates
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Set x-axis labels
ax.set_xticks(dates)
ax.set_xticklabels(dates, rotation=45, ha="right", fontsize=12)

# Set y-axis labels
ax.set_ylabel('Ratio', fontweight='bold', fontsize=14, color="dimgray")

# Remove unnecessary gridlines and axis outlines
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Plot BEP ratio bars with annotations and violet color
ax.bar(
    dates - bar_spacing,
    BEP_ratio,
    width,
    label='BEP Ratio',
    color='#8F00FF',  # Violet
    alpha=0.8,
    linewidth=1,
    capstyle='round',
)
for i, (date, value) in enumerate(zip(dates, BEP_ratio)):
    ax.text(date - bar_spacing, value + 0.05, f"{value:.2f}", ha="center", va="bottom", fontsize=12)

# Plot ROA ratio bars with offset and annotations and blue color
ax.bar(
    dates,
    ROA_ratio,
    width,
    label='ROA Ratio',
    color='#0072B2',  # Blue
    alpha=0.8,
    linewidth=1,
    capstyle='round',
)
for i, (date, value) in enumerate(zip(dates, ROA_ratio)):
    ax.text(date, value + 0.05, f"{value:.2f}", ha="center", va="bottom", fontsize=12)

# Plot ROE ratio bars with offset and annotations and teal color
ax.bar(
    dates + bar_spacing,
    ROE_ratio,
    width,
    label='ROE Ratio',
    color='#008080',  # Teal
    alpha=0.8,
    linewidth=1,
    capstyle='round',
)
for i, (date, value) in enumerate(zip(dates, ROE_ratio)):
    ax.text(date + bar_spacing, value + 0.05, f"{value:.2f}", ha="center", va="bottom", fontsize=12)

# Add legend
legend = ax.legend(loc='upper right', frameon=True, title='Ratios', title_fontsize=12, fancybox=True)

# set title
plt.title("**PROFIT RATIOS**", fontsize=18, fontweight="bold",color='seagreen')

# Plot line with desired color and style
ax.plot([min_date, max_date], [y_start, y_end], color='seagreen', linewidth=2, linestyle='--')

# Tight layout and show plot
plt.tight_layout()
plt.show()

```

```python
# Set a seaborn style
sns.set(style="whitegrid")

# Plot the bar graph
ax = sai_ratios.plot.bar(x='Date', y='Debt Ratio', color='skyblue', edgecolor='black', linewidth=1.2)

# Annotate bars with values 
for bar in ax.patches:
    ax.annotate(f'{bar.get_height():.2f}', 
                (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', 
                fontsize=12, color='black')

# Adding labels and title
plt.ylabel("Debt Ratio", fontsize=12)
plt.title("DEBT RATIO", fontsize=16,weight="bold")

# Rotating x-axis labels for better readability (optional)
plt.xticks(rotation=45, ha='right')


# Add a legend (optional)
plt.legend(["Debt Ratio"], loc='upper left')


# Remove the top and right spines for aesthetics
sns.despine()

# Show the plot
plt.show()
```
