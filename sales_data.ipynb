##### We are going to plot month on x-axis since it's a time series and time is independent and sales on y-axis since it's dependent on time which is month in this case

```python
fig,(ax1,ax2,ax3,ax4)=plt.subplots(4,1,figsize=(6,12)) 
ax1.plot(monthly_sales["Months"],monthly_sales["sales"])
```

### since we have to plot the same line graph on all 4 graph we will use loop instead of plotting them one by one

```python
fig,(ax1,ax2,ax3,ax4)=plt.subplots(4,1,figsize=(6,12))
axes = [ax1,ax2,ax3,ax4]
for i in axes:
    i.plot(monthly_sales["Months"],monthly_sales["sales"])
```

```python
fig,(ax1,ax2,ax3,ax4)=plt.subplots(4,1,figsize=(6,12))
axes = [ax1,ax2,ax3,ax4]
for i in axes:
    i.plot(monthly_sales["Months"],monthly_sales["sales"])
    i.set_yticklabels([])
    i.set_yticks([])
    i.set_xticklabels([])
    i.set_xticks([])
#     i.tick_params(bottom=False, left=False)
```

```python
fig,(ax1,ax2,ax3,ax4)=plt.subplots(4,1,figsize=(6,12))
axes = [ax1,ax2,ax3,ax4]
ax1.tick_params(bottom=0, left=0)
for i in axes:
    i.plot(monthly_sales["Months"],monthly_sales["sales"])
    i.set_yticklabels([])
    i.set_yticks([])
    i.set_xticklabels([])
    i.set_xticks([])# we could also  use i.tick_params(bottom=False, left=False
    # Removing params
    spines_params = ["left","right","top","bottom"]
    for j in spines_params:
        i.spines[j].set_visible(False)
```
